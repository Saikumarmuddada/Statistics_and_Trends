# Relational Plot

```python
#Line Plot: Loan Interest Rate vs. Age
plt.figure(figsize=(8,5))
sns.lineplot(x=data['person_age'], y=data['loan_int_rate'])
plt.title('Loan Interest Rate vs. Age')
plt.xlabel('Age')
plt.ylabel('Loan Interest Rate')
plt.show()
```

```python
#Step Plot: Credit Score vs. Employment Experience
plt.figure(figsize=(8,5))
plt.step(x='person_emp_exp', y='credit_score', where='mid', linestyle='-', marker='o', data=data)
plt.title('Credit Score vs. Employment Experience')
plt.xlabel('Employment Experience')
plt.ylabel('Credit Score')
plt.show()
```

```python
# Scatter Plot: Loan Amount vs. Loan Percent Income
plt.figure(figsize=(8,5))
sns.scatterplot(x='loan_amnt', y='loan_percent_income', hue='loan_status', data=data)
plt.title('Loan Amount vs. Loan Percent Income')
plt.xlabel('Loan Amount')
plt.ylabel('Loan Percent Income')
plt.show()
```

# Categorical Plots

```python
# Histogram: Credit Score Distribution
plt.figure(figsize=(8,5))
sns.histplot(data['credit_score'], bins=30, kde=True)
plt.title('Distribution of Credit Scores')
plt.xlabel('Credit Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar Plot: Loan Intent Count
plt.figure(figsize=(8,5))
sns.countplot(x=data['loan_intent'])
plt.title('Number of Loans by Intent')
plt.xlabel('Loan Intent')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Horizontal Bar Plot: Loan Status by Home Ownership
plt.figure(figsize=(8,5))
data.groupby('person_home_ownership')['loan_status'].value_counts().unstack().plot(kind='barh', stacked=True)
plt.title('Loan Status by Home Ownership')
plt.xlabel('Count')
plt.ylabel('Home Ownership')
plt.legend(title='Loan Status')
plt.show()
```

```python
# Pie Chart: Education Level Distribution

plt.figure(figsize=(6,6))
data['person_education'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm')
plt.title('Education Level Distribution')
plt.ylabel('')
plt.show()
```

# Statistical Plots

```python
#Violin Plot: Credit History Length by Gender
plt.figure(figsize=(8,5))
sns.violinplot(x=data['person_gender'], y=data['cb_person_cred_hist_length'])
plt.title('Credit History Length by Gender')
plt.xlabel('Gender')
plt.ylabel('Credit History Length')
plt.show()
```

```python
#Pair Plot

sns.pairplot(data=data, hue='loan_status')
plt.show()
```

```python
data.head()
```

```python
# convert all categories into numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['person_home_ownership'] = le.fit_transform(data['person_home_ownership'])
```

```python
data['person_gender'] = le.fit_transform(data['person_gender'])
data['loan_intent'] = le.fit_transform(data['loan_intent'])
```

```python
data['person_education'] = le.fit_transform(data['person_education'])
data['previous_loan_defaults_on_file'] = le.fit_transform(data['previous_loan_defaults_on_file'])
```

```python
# Correlation Heatmap


plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Box Plot: Loan Interest Rate by Education Level

plt.figure(figsize=(8,5))
sns.boxplot(x='person_education', y='loan_int_rate', data=data)
plt.title('Loan Interest Rate by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Loan Interest Rate')
plt.xticks(rotation=45)
plt.show()
```

# Statistical Moments
### Mean, Variance, Skewness, and Kurtosis

```python
from scipy.stats import skew, kurtosis

# Mean
mean_values = data.mean()

# Variance
variance_values = data.var()
# Skewness
skew_values = data.apply(skew)

# Kurtosis
kurtosis_values = data.apply(kurtosis)

# Print Results
print("Mean:\n", mean_values)
print("\nVariance:\n", variance_values)
print("\nSkewness:\n", skew_values)
print("\nKurtosis:\n", kurtosis_values)
```
